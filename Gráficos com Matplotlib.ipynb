# Aula 3 - Anotações e diversos gráficos

```python
# Figura geral
fig = plt.figure(figsize=(10,6))
eixo1 = fig.add_axes([0,0,1,1])

# Gráfico
eixo1.grid(True)
eixo1.plot(dados['data'], dados['temperatura'], color='sandybrown')
eixo1.set_title('Temperatura no momento', fontsize=18, pad=20)
eixo1.set_xlabel('Data', fontsize=14)
eixo1.set_ylabel('Temperatura (K)', fontsize=14)

# Limitantes superior e inferior
eixo1.axhline(max(dados['temperatura']), color='k', linestyle='--')
eixo1.axhline(min(dados['temperatura']), color='k', linestyle='--')
```

```python
# Figura geral
fig = plt.figure(figsize=(10,6))
eixo1 = fig.add_axes([0,0,1,1])

# Gráfico
eixo1.grid(True)
eixo1.plot(dados['data'], dados['temperatura'], color='sandybrown')
eixo1.set_title('Temperatura no momento', fontsize=18, pad=20)
eixo1.set_xlabel('Data', fontsize=14)
eixo1.set_ylabel('Temperatura (K)', fontsize=14)

# Anotação de ponto de máximo
xMaxOrigin = dados['data'][dados['temperatura'].idxmax()]
yMaxOrigin = max(dados['temperatura'])
xMaxDestin = dados['data'][dados['temperatura'].idxmax() - 7000]
yMaxDestin = max(dados['temperatura']) - 5
eixo1.annotate('Máximo', xy=(xMaxOrigin, yMaxOrigin),
               xytext=(xMaxDestin, yMaxDestin), fontsize=14,
               arrowprops=dict(facecolor='k'))

# Limitantes superior e inferior
eixo1.axhline(max(dados['temperatura']), color='k', linestyle='--')
eixo1.axhline(min(dados['temperatura']), color='k', linestyle='--')
```
